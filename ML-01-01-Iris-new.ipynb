### Предсказание

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train,X_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=12)

knk=KNeighborsClassifier() # классификатор методом ближайших соседей
knk.fit(X_train,y_train) #подбираем соответствия между тренировочными данными Х и сопоставляем их с тренировочными Y, получаем завичимость
#по умолчанию эвклидова метрика, weights='uniform' - не штрафует за большое расстояние
#n_neighbors=5 - пять ближайших соседей
#fit записал в knk алгоритм
```

```python
y_pred=knk.predict(X_test) #предсказывает результаты с помощью тестового набора, можем сравнить у предсказанный с у истиным
print('Правильность на тестовом наборе: {:.3f}'.format(np.mean(y_pred==y_test)))
```

```python
knk_train=knk.score(X_train , y_train) #score считает метрику правильности полученных результатов (то есть насколько правильно предсказывает модель)
knk_test=knk.score(X_test , y_test)
print("тренировочный: {:.3f}".format(knk_train))
print("тестовый: {:.3f}".format(knk_test))
```

### Кросс-валидация

```python
from sklearn.model_selection import cross_val_score 

knk = KNeighborsClassifier()
cross_vall = cross_val_score(knk, iris_dataset.data, iris_dataset.target, cv=5)
#cv делит выборку на 5, то есть, если у нас было 150 наблюдений, 30 уйдет в тест и 120 в тренировку
#плохо, потому что мы учим модель на упорядоченных данных

np.set_printoptions(precision=4)
print("Значения правильности перекрестной проверки:",cross_vall)
```

```python
from sklearn.model_selection import ShuffleSplit

knk = KNeighborsClassifier()
shuffle_split = ShuffleSplit(test_size=.3, train_size=.6, n_splits=5, random_state=12)
# здесь мы взяли 30 в тест, 60 в тренировку и 10 потеряли (все это случайным образом)
#затем повторяем этот алгоритм
# случайный образ необходим, так как если мы делим по упорядоченным данным, то в тренировочную выборку могут вообще не попасть некоторые классы

cross_vall = cross_val_score(knk, iris_dataset.data, iris_dataset.target, cv=shuffle_split)

np.set_printoptions(precision=4)
print("Значения правильности перекрестной проверки:",cross_vall)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

knk = KNeighborsClassifier()
strat_shuffle_split = StratifiedShuffleSplit(test_size=.3, train_size=.6, n_splits=5, random_state=12)

cross_vall = cross_val_score(knk, iris_dataset.data, iris_dataset.target, cv=strat_shuffle_split)

np.set_printoptions(precision=4)
print("Значения правильности перекрестной проверки:",cross_vall)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
matr = []
cross = []
n = [1,2,3,4,5,6]
p = [1,2,3,6,10]
for i in n:
    cross = []
    for j in p:
        knk = KNeighborsClassifier(p=j, n_neighbors=i)
        strat_shuffle_split = StratifiedShuffleSplit(test_size=.3, train_size=.6, n_splits=5, random_state=10)
        cross_vall = cross_val_score(knk, iris_dataset.data, iris_dataset.target, cv=strat_shuffle_split)
        cross.append(np.mean(cross_vall))
    matr.append(cross)
```

```python
sns.heatmap(matr, linewidths=.005, xticklabels = p, yticklabels = n, annot=True)
```
