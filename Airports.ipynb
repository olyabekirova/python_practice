```python
routes = pd.read_csv('routes.csv', names = ['Airline', 'AirlineID', 'SourceAirport', 'Source Airport ID', 'DestinationAirport', 'Destination Airport ID', 'Codeshare', 'Stops', 'Equipment' ])
route = pd.DataFrame (routes)

#display(route)
x=["BPX",'NGQ']

a=route[route.SourceAirport.isin(x)].AirlineID
c=route[route.DestinationAirport.isin(x)].AirlineID

id = np.concatenate((a,c))
np.unique(id)                                               
```
